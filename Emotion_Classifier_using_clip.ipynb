```python
from PIL import Image
import requests
from transformers import AutoProcessor, CLIPModel
import numpy as np
```

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = AutoProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
def get_prob(path):

    # image = Image.open(requests.get(url, stream=True).raw)
    image = Image.open(path)

    emotions_list = ['Joy',
        'Interest',
        'Serenity',
        'Anticipation',
        'Vigilance',
        'Apprehension',
        'Acceptance',
        'Trust',
        'Sadness',
        'Ecstasy',
        'Admiration',
        'Annoyance',
        'Anger',
        'Grief',
        'Pensiveness',
        'Fear',
        'Boredom',
        'Rage',
        'Distraction',
        'Terror',
        'Surprise',
        'Disgust',
        'Amazement',
        'Loathing'
    ]

    inputs = processor(
        text=emotions_list, images=image, return_tensors="pt", padding=True
    )
    outputs = model(**inputs)
    logits_per_image = outputs.logits_per_image
    probs = logits_per_image.softmax(dim=1).tolist()

    d = {}
    for i in range(len(probs[0])):
        d[emotions_list[i]] = probs[0][i]

    sorted_d = sorted(d.items(), key=lambda kv: -kv[1])
    return sorted_d
```

```python
def find_percent(emo_list, emotion):
    emo_dict = {}
    for emo in emo_list:
        emo_dict[emo[0]] = emo[1]
    return emo_dict[emotion]
```

```python
url = '//mnt/nas/swethamagesh/emotion/GANimation/sample_dataset/imgs/anger_00003.jpg'
emo_list = get_prob(url)
print("anger", find_percent(emo_list, 'Anger'))
```
